**2 Given the list [10, 15, 20, 25, 30], write a function that returns the square of only even numbers**

````python
def squares(my_list):
  squares_even = []
  for number in my_list:
    if number%2 == 0:
      squares_even.append(number**2)
  return squares_even

my_list = [10, 15, 20, 25, 30]
squares(my_list)
````

**7.What does the .groupby() function do in pandas? Show an example using dummy data.** **bold text**

groupby() function used to splits data into groups based on column values

````python
import pandas as pd

# Create dummy data
data = {'car': ['BMW', 'Mercedes', 'TOYOTA', 'Jaguar', 'Ferari', 'Audi'],
        'color': ['red', 'black', 'white', 'black', 'white', 'red'],
        'price': [500000., 600000, 650000, 450000, 550000, 700000]}
df = pd.DataFrame(data)

# Group by 'car' and calculate the average price for each car
car_prices = df.groupby('car')['price'].mean()

#To see the output, run the code.
````
