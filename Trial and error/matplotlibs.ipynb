```python
import numpy as np
import matplotlib.pyplot as plt

# Data
X = np.array([1, 2, 3, 4, 5])
y = np.array([1, 2, 1.3, 3.75, 2.25])

# Add bias
X_mat = np.vstack([np.ones(len(X)), X]).T

def lwr(x_query, X, y, tau):
    m = len(X)
    X_mat = np.vstack([np.ones(m), X]).T
    x_query_mat = np.array([1, x_query])
    
    # compute weights
    W = np.diag(np.exp(-((X - x_query) ** 2) / (2 * tau ** 2)))
    
    # normal equation with weights
    theta = np.linalg.pinv(X_mat.T @ W @ X_mat) @ (X_mat.T @ W @ y)
    
    return x_query_mat @ theta

# Predict for multiple x values
x_vals = np.linspace(1, 5, 100)
y_preds = [lwr(x, X, y, tau=0.5) for x in x_vals]

# Plot
plt.scatter(X, y, color='red', label="Training data")
plt.plot(x_vals, y_preds, label="LWR prediction")
plt.title("Locally Weighted Regression")
plt.legend()
plt.show()
```
