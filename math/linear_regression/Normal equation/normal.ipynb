#### The normal equation provides an analytical solution comuting the optimal parameters in single step

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
x = 3*np.random.rand(50)
y = 2 + 3*x + np.random.rand(50)
```

```python
plt.plot(x,y ,"o")
```

```python
x_b = np.c_[np.ones((50,1)), x]
```

```python
#Normal_Equation 
beta = np.linalg.inv(x_b.T.dot(x_b)).dot(x_b.T).dot(y)
beta
```

```python
#predict values for y 
x_predict = np.array([[0], [3]])
```

```python
x_predict_b = np.c_[np.ones((2, 1)), x_predict]
x_predict_b
```

```python
y_predict = x_predict_b.dot(beta)
```

```python
plt.plot(x, y, 'o')
plt.plot(x_predict, y_predict, 'r-')
```
